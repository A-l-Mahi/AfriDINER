```python
import os
import pandas as pd
```

```python
base_path = '../metrics'
dfs_list = os.listdir(base_path)

afro = {}
xlmr = {}

for i in dfs_list:
    if i.endswith('.csv') and "val" in i:
        df = pd.read_csv(os.path.join(base_path, i))
        if 'Afro' in i:
            if "baseline" in i:
                afro["baseline"] = df
            elif "counterfactual" in i:
                afro["counterfactual"] = df
        elif 'XLMr' in i:
            if "baseline" in i:
                xlmr["baseline"] = df
            elif "counterfactual" in i:
                xlmr["counterfactual"] = df


```

```python
print(afro.keys(), xlmr.keys())
```

```python
# Direct comparison between XLM-R baseline and Afro-XLM-R counterfactual

import matplotlib.pyplot as plt


plt.figure(figsize=(15, 5))

# Subplot 1: Accuracy comparison
plt.subplot(1, 3, 1)
plt.plot(afro['counterfactual']['epoch'], afro['counterfactual']['acc'], label='CF', marker='o', alpha=0.7)
plt.plot(afro['baseline']['epoch'], afro['baseline']['acc'], label='base', marker='s', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 2: F1 Score comparison
plt.subplot(1, 3, 2)
plt.plot(afro['counterfactual']['epoch'], xlmr['counterfactual']['f1'], label='CF', marker='o', alpha=0.7)
plt.plot(afro['baseline']['epoch'], afro['baseline']['f1'], label='base', marker='s', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score Comparison')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 3: Loss comparison
plt.subplot(1, 3, 3)
plt.plot(afro['counterfactual']['epoch'], afro['counterfactual']['loss'], label='CF', marker='o', alpha=0.7)
plt.plot(afro['baseline']['epoch'], afro['baseline']['loss'], label='base', marker='s', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Comparison')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
